```python
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
data = np.load('bunny.npy')
data
```

```python
data.shape
```

```python
image_size = 30
som = MiniSom(image_size, image_size, 3, sigma=1,
              learning_rate=.25,)



```

```python
weights = som.get_weights()

x = weights[:,:,0].reshape(1,-1).transpose()
y = weights[:,:,1].reshape(1,-1).transpose()
z = weights[:,:,2].reshape(1,-1).transpose()

fig = plt.figure()
ax = plt.axes(projection='3d' )
ax.scatter3D(x, y ,z ,alpha=0.1 ,edgecolor='blue');
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

```python
som.train(data, 250000 ,random_order=True, verbose=True)
```
